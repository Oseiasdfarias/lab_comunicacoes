```python
from __future__ import division
import numpy as np
import math
import matplotlib.pyplot as plt
```

**Exemplo 02 - Sinal Senoidal** \\

Queremos mostrar e analisar 5 ciclos completos da onda senoidal com f = 0.5, correspondendo a uma duração total de 10 segundos.
A onda senoidal é amostrada a uma taxa: \\
1ª - fs = 1Hz (saltos 1 seg) -> 10 × fs = 10 Hz pontos de dados. \\
2ª - fs = 2Hz (saltos 0.5 seg) -> 10 × fs = 20 Hz pontos de dados. \\
3ª - fs = 10Hz (saltos 0.1 seg) -> 10 × fs = 100 Hz pontos de dados.

```python
# Variaveis Inicial ====================================
dt = 0.1 #dx (Passo/Salto) = 1s => (dt = 1/fs), fs é a Taxa de Amostragem do Sinal 
t = np.arange(0,10, dt) #Eixo t - tempo em segundo | intervalo [0-10] | com dt (salto) = 1s 

f = 0.5  #frequência da portadora
A = 1.0  #amplitude da portadora

# Calculando os valores do Sinal - s(t)===================
s = A * np.cos(2 * np.pi * f * t)

# Plot =================================================
fig = plt.figure(figsize=(10,5))

cx = fig.add_subplot(1, 1, 1)
cx.set_title('Sinal - Domínio do Tempo')
cx.set_xlabel('Tempo (s)')
cx.set_ylabel('Amplitude - Sinal')
cx.plot(t, s, ".")
cx.plot(t, s)

plt.show()
```

```python
F = np.fft.fft(s)  
freq = np.fft.fftfreq(np.size(F), d=dt)

# Plot =================================================
fig = plt.figure(figsize=(10,5))

cx = fig.add_subplot(1, 1, 1)
cx.set_title('Tranfomada de Fourie do Sinal - Domínio da Frequência')
cx.set_xlabel('Frequência (Hz)')
cx.set_ylabel('Amplitude')
cx.plot(freq, np.abs(F), ".")
cx.plot(freq, np.abs(F))

plt.show()
```

**Exemplo 03 - Sinal Senoidal** \\

Queremos mostrar e analisar 10 segundos de um sinal senoidal, com frequência. O sinal é amostrada a uma taxa: \\
1ª - fs = 100Hz (saltos 0.01 seg) -> 10 × fs = 1000 Hz pontos de dados. \\

```python
# Variaveis Inicial ====================================
dt = 0.01 #dx (Passo/Salto) = 1s => (dt = 1/fs), fs é a Taxa de Amostragem do Sinal
t = np.arange(0,10, dt) #Eixo t - tempo em segundo | intervalo [0-10] | com dt (salto) = 1s 

f = 0.5  #frequência do sinal
A = 1  #amplitude do sinal

# Calculando os valores do Sinal ===================
s = A * np.cos(2 * np.pi * f * t)

# Plot =================================================
fig = plt.figure(figsize=(10,5))

cx = fig.add_subplot(1, 1, 1)
cx.set_title('Sinal - Domínio do Tempo')
cx.set_xlabel('t')
cx.set_ylabel('s(t)')
cx.plot(t, s, ".")
cx.plot(t, s)

plt.show()
```

```python
F = np.fft.fft(s)  
freq = np.fft.fftfreq(np.size(F), d=dt)

# Plot =================================================
fig = plt.figure(figsize=(10,5))

cx = fig.add_subplot(1, 1, 1)
cx.set_title('Tranfomada de Fourie do Sinal Senoidal')
cx.set_xlabel('Frequência (Hz)')
cx.set_ylabel('Amplitude')
cx.plot(freq, np.abs(F), ".")
cx.plot(freq, np.abs(F))

plt.show()
```
